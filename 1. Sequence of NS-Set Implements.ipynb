## A.	The Maji's Algorithm: From Set to Sequence

```python
import numpy as np
from maji_nss import *
######
at1 = [[0.6, 0.3, 0.1],[0.2, 0.3, 0.4],[0.9, 0.6, 0.7]]
bt1 = [[0.1, 0.4, 0.6],[0.8, 0.3, 0.6],[0.2, 0.6, 0.9]]
ct1 = [[0.4, 0.6, 0.2],[0.3, 0.5, 0.2],[0.6, 0.9, 0.6]]
dt1 = [[0.3, 0.3, 0.4],[0.1, 0.7, 0.6],[0.2, 0.6, 0.4]]
######
mat = np.array([at1,bt1,ct1,dt1])
######
s = [4,3]
######
print("The Maji’s Algorithm for NS-Set in Decision-Making Problem")
print("")
print("> Step 1. Comparison Matrix")
print(matrix(s,mat))
print("")
print("> Step 2. Object Scores")
print(score(s,mat))
print("")
print("> Step 3. Output")
print(maji(s,mat))
```
